```python
import numpy as np
import matplotlib.pyplot as plt

def load_data(filename):
    sqft = []
    price = []
    with open(filename, "r") as f:
        next(f)
        for line in f:
            s, p = line.strip().split(",")
            sqft.append(float(s))
            price.append(float(p))
    return np.array(sqft), np.array(price)

def gradient_descent(x, y, lr=0.00000001, epochs=5000):
    m = 0
    b = 0
    n = len(x)

    for _ in range(epochs):
        y_pred = m * x + b

        dm = (-2/n) * np.sum(x * (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)

        m -= lr * dm
        b -= lr * db

    return m, b

def predict(x, m, b):
    return m * x + b

def plot_best_fit(x, y, m, b):
    plt.scatter(x, y, label="Data Points")

    x_line = np.linspace(min(x), max(x), 100)
    y_line = m * x_line + b

    plt.plot(x_line, y_line, color="green", label="Best Fit Line (GD)")
    plt.xlabel("Square Feet")
    plt.ylabel("Price")
    plt.legend()
    plt.title("Gradient Descent Linear Regression (Manual)")
    plt.show()

if __name__ == "__main__":
    x, y = load_data("housing_data.csv")
    m, b = gradient_descent(x, y)

    price_2500 = predict(2500, m, b)

    print(f"Gradient Descent Prediction for 2500 sq ft = {price_2500:.2f}")

    plot_best_fit(x, y, m, b)
```
